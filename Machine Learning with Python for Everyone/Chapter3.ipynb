```python
import pandas as pd
from sklearn.datasets import load_iris
import sklearn as sk
from sklearn.metrics import confusion_matrix
```

```python
iris=load_iris()
```

```python
iris_df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
iris_df['Target']=iris.target
```

```python
iris_df.head()
```

```python
(iris_train_ftrs, iris_test_ftrs,
iris_train_tgt, iris_test_tgt) = sk.model_selection.train_test_split(iris.data,
iris.target,
test_size=.25)
```

```python
iris_train_ftrs.shape, iris_test_ftrs.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
param_grid={'n_neighbors':[1,3,10,20]}
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(knn, param_grid,cv=5)
grid.fit(iris_train_ftrs,iris_train_tgt)
```

```python
best_params=grid.best_params_
```

```python
knn1=KNeighborsClassifier(**best_params)
knn1.fit(iris_train_ftrs,iris_train_tgt)
predictions=knn1.predict(iris_test_ftrs)
accuracy=knn1.score(iris_test_ftrs,iris_test_tgt)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(iris_test_tgt,predictions)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_items={'knn':KNeighborsClassifier(n_neighbors=3),'nb':GaussianNB()}
for name,model1 in model_items.items():
    model1.fit(iris_train_ftrs,iris_train_tgt)
    predictions=model1.predict(iris_test_ftrs)
    accuracy=model1.score(iris_test_ftrs,iris_test_tgt)
    print(name,accuracy)
```
